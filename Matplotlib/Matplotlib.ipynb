### Matplotlib example workflow

```python
# 1. Prepare Data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup Plot
fig, ax = plt.subplots(figsize=(10,10)) # width, height

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title = "Simple Plot",
       xlabel = "X-axis",
       ylabel = "Y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

### Two options for subplots

```python
# subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize=(10,5))

# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize =(10,5))

# plot to each different index
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

### which one should i use? (pyplot vs matplotlib OO method?)
 
* When plotting something quickly, i have to use pyplot method
* When plotting something more advanced, then i should use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind="scatter",
            x = 'age',
            y = 'chol',
            c = 'target');
```

```python
# OO method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax);
```

```python
# OO method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x = 'age',
             y = 'chol',
             c = 'target',
             ax = ax)
ax.set_xlim([45,100]);
```

```python
## OO (object oriented) method form scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50['age'],
                    y =over_50['chol'],
                    c = over_50['target'])

# Customize the plot
ax.set(title = "Heart Disease and cholesterol Levels",
      xlabel = "Age",
      ylabel = "Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title= "Target")

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
over_50.head()
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize=(10,10))

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'])

# Customize ax0
ax0.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title= "Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle = '--');

#---------------------------2nd plot-----------
#### Add data to ax1
scatter = ax1.scatter(x= over_50['age'],
                     y = over_50['thalach'],
                     c = over_50['target'])

# Customize ax1
ax1.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Max Heart Rate");

# add a legend to ax1
ax1.legend(*scatter.legend_elements(), title= "Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle = '--');
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize=(10,10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'])

# Customize ax0
ax0.set(title = "Heart Disease and Cholesterol Levels",
       ylabel = "Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title= "Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle = '--');

#---------------------------2nd plot-----------
#### Add data to ax1
scatter = ax1.scatter(x= over_50['age'],
                     y = over_50['thalach'],
                     c = over_50['target'])

# Customize ax1
ax1.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Max Heart Rate");

# add a legend to ax1
ax1.legend(*scatter.legend_elements(), title= "Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle = '--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight = "bold");
```

```python
 # Set the style
plt.style.use('seaborn-v0_8-whitegrid')

## OO (object oriented) method form scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50['age'],
                    y =over_50['chol'],
                    c = over_50['target'],
                    cmap = "winter") # this changes the colour scheme

# Customize the plot
ax.set(title = "Heart Disease and cholesterol Levels",
      xlabel = "Age",
      ylabel = "Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title= "Target")

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize=(10,10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'],
                     cmap = 'winter')

# Customize ax0
ax0.set(title = "Heart Disease and Cholesterol Levels",
       ylabel = "Cholesterol")

# Change the x axix limits
ax0.set_xlim([50,80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title= "Target")

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle = '--', color = 'red');

#---------------------------2nd plot-----------
#### Add data to ax1
scatter = ax1.scatter(x= over_50['age'],
                     y = over_50['thalach'],
                     c = over_50['target'],
                     cmap = 'winter')

# Customize ax1
ax1.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Max Heart Rate");

# Change the ax1 axix limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# add a legend to ax1
ax1.legend(*scatter.legend_elements(), title= "Target")

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle = '--', color ='red');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight = "bold");
```
